```python
# Project 5: Predictive Regression Architecture (Movie Revenue Predictor)
# STEP 1: Load foundational regression modeling and processing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("[SYSTEM NOTICE] Regression modeling modules successfully initialized.")
```

```python
# STEP 2: Generate a synthesized entertainment industry dataset
np.random.seed(50)
num_movies = 150

# Generating core continuous attributes (in Millions of Dollars)
production_budget = np.random.uniform(5.0, 150.0, size=num_movies)
marketing_spend = production_budget * np.random.uniform(0.2, 0.6, size=num_movies)
social_buzz_score = np.random.randint(10, 100, size=num_movies)

# Define mathematical linear relationship with random market noise to simulate real revenue
box_office_revenue = (production_budget * 1.5) + (marketing_spend * 2.1) + (social_buzz_score * 0.3) + np.random.normal(0, 10, size=num_movies)

# Enforce positive values for realistic financials
box_office_revenue = np.clip(box_office_revenue, 1.0, None)

# Assemble features into a structured box office dataframe
movie_data = pd.DataFrame({
    'movie_id': range(5001, 5001 + num_movies),
    'production_budget_M': np.round(production_budget, 2),
    'marketing_spend_M': np.round(marketing_spend, 2),
    'social_buzz': social_buzz_score,
    'box_office_revenue_M': np.round(box_office_revenue, 2)
})

# Save to disk as raw storage backup
movie_data.to_csv('movie_revenue_data.csv', index=False)
print("=" * 60)
print("   BOX OFFICE REPOSITORY GENERATION COMPLETE")
print("=" * 60)
print(movie_data.head(10).to_string(index=False))
```

```python
# STEP 3: Segregate predictive variables from targets and split into training slices
X = movie_data[['production_budget_M', 'marketing_spend_M', 'social_buzz']]
y = movie_data['box_office_revenue_M']

# Split data: 80% for training the weights, 20% for continuous evaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=50)

# Normalize spatial distribution to scale variables perfectly
feature_scaler = StandardScaler()
X_train_scaled = feature_scaler.fit_transform(X_train)
X_test_scaled = feature_scaler.transform(X_test)

print(f"[Preprocessing Success]: Training Dimension: {X_train_scaled.shape} | Evaluation Dimension: {X_test_scaled.shape}")
```

```python
# STEP 4: Instantiate the Ordinary Least Squares Linear Regression model and fit data
print("=" * 60)
print("     TRAINING PHASE: FITTING ORDINARY LEAST SQUARES MODEL")
print("=" * 60)

regression_engine = LinearRegression()
regression_engine.fit(X_train_scaled, y_train)

# Predict values on the unseen test matrix
revenue_predictions = regression_engine.predict(X_test_scaled)

print("[Model Execution Notification]: Linear weights optimized and predictions saved.")
```

```python
# STEP 5: Calculate error margins using continuous regression metrics
print("=" * 60)
print("   STATISTICAL MODEL DIAGNOSTICS & LOSS METRICS")
print("=" * 60)

# Calculate performance metrics
mae = mean_absolute_error(y_test, revenue_predictions)
mse = mean_squared_error(y_test, revenue_predictions)
r2 = r2_score(y_test, revenue_predictions)

print(f"Mean Absolute Error (MAE) : ${mae:.2f} Million")
print(f"Mean Squared Error (MSE)  : ${mse:.2f} Million")
print(f"R-squared ($R^2$) Score      : {r2:.4f} (Variance Explanation Limit)")

print("\n--- Side-by-Side Actual vs Predicted Projections ---")
comparison_df = pd.DataFrame({
    'Actual Revenue ($M)': y_test.values,
    'Predicted Revenue ($M)': np.round(revenue_predictions, 2)
}).head(10)
print(comparison_df.to_string(index=False))
```
